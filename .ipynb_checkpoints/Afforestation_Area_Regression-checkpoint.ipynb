## 1.3. Data cleaning

### I'll delete the rows that contain 'Total Afforestation' since I want to analyze the plantations carried out for each specie and forest owner instead of analyzing them in their entirety, this is to avoid generating redundancy or noise

```python
filtro = (afforestation_df['Species'] != 'Total Afforestation') & (afforestation_df['Forest Owner'] != 'Total Afforestation')    # Remove rows where either the 'Species' or 'Forest Owner' is 'Total Afforestation'
afforestation_df = afforestation_df[filtro]   # Update the DataFrame with the requested conditions 
afforestation_df                              
```

### I'll remove columns 'Statistic Label' and 'UNIT' due to they both have the same value: 'Afforestation Area' and 'Hectares' respectively

```python
afforestation_df.drop(['Statistic Label', 'UNIT'], axis=1, inplace=True)   # axis=1 (remove columns instead of rows), inplace=True (modify the DataFrame in place)
```

### We'll rename the columns 'VALUE' and 'Forest Owner'

```python
column_dictionary = {'VALUE':'Value_ha', 'Forest Owner':'Forest_Owner'}         # Creation of a dictionary
afforestation_df.rename(columns=column_dictionary, inplace=True)                # Method to rename the column named 'VALUE' to 'Value_ha'
afforestation_df                                                                # display the DataFrame to see the changes made
```

```python
afforestation_df.info() 
```

### We can also see that the data type of the columns is correct and we don't need to perform a conversion

```python
# Indexes don't start from zero and there are 2592 entries which start from 5 to 5183 
# Let's update the indexes, so that the information is ordered correctly
afforestation_df.reset_index(drop=True, inplace=True)   # Reset indexes to start from 0, drop=True (clear all indexes), inplace=True(replace the new indexes in the DataFrame)
afforestation_df.info(verbose=False)                    # verbose=False (it only will display range index, since it displays specific information)
```

### 2.1.2. Transformation of Variables

```python
# FunctionTransformer transforms real values to transformed values using logarithmic function
# To return to the initial values, we transform the logarithmic function into an exponential function (inversely proportional)
from sklearn.preprocessing import FunctionTransformer
transformer = FunctionTransformer(np.log1p, validate=True)  # np.log1p (creates a transformer object that applies the 'log1p' function from the NumPy library, 'validate=True' (argument ensures that the transformation is validated)
transformed_values = transformer.transform(afforestation_df.select_dtypes(exclude=['object']))  # Applies the 'transformer' to the 'afforestation_df' DataFrame, transforming the numerical columns (excluding object data types)
```

```python
columns_transformed_values = afforestation_df.select_dtypes(exclude=['object']).columns
transformed_df = pd.concat([afforestation_df.select_dtypes(include=['object']), pd.DataFrame(transformed_values, columns=columns_transformed_values)], axis=1)
transformed_df.head()
```

```python
# After the transformation

sns.heatmap(transformed_df.corr(), annot = True, fmt = '.2f', linewidths = 2)
ax.set_title('Correlation between Variables - Transformation Log', fontsize=20)
```

```python
# After the transformation

transformed_correlation = transformed_df.corr()
mask = np.zeros_like(transformed_correlation)
mask[np.triu_indices_from(mask)] =True                  # Mask to see some points of the table
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(13, 8))               # Save context to ax
    ax = sns.heatmap(transformed_correlation, cmap=cores, mask=mask, square=True, fmt='.2f', annot=True)
    ax.set_title('Correlation between Variables - Transformation Log', fontsize=20)
```

```python
# Convert to Symetric Distribution

ax = sns.histplot(data=transformed_df, x='Value_ha', kde=True)
ax.figure.set_size_inches(20,10)
ax.set_title('Histogram of transformed afforestation (Value_ha)', fontsize = 20)
ax.set_xlabel('Value_ha', fontsize = 15)
ax.set_ylabel('Frequency', fontsize = 15)
```

## 2.2. Linear Regression

"Regression analysis refers to the study of the dependence of a variable, the dependent variable, in relation to one or more variables, the explanatory variables, with the objective of estimating and/or predicting the mean value (of the population) of the first in terms of the known or fixed values (in repeated sampling) of the latter.”

- Simple regression: the dependent variable (y) is related to a single explanatory variable (X).
- Multiple regression: The dependent variable (y) is related to more than a single explanatory variable (Xi).

## 2.3. Adjustment and prediction

### 2.3.1. Data division

```python
# Independent Variables
X = afforestation_dummies_df.drop(['Value_ha'], axis=1)

# Dependent Variables
y = afforestation_dummies_df['Value_ha']       # Variable that I'm going to predict
```

```python
y
```

```python
from sklearn.model_selection import train_test_split      
```

```python
# Splitting the data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)     # test_size=0.3 (use 30% of the data for testing and 70% for training), random_state=42(set a specific value for random_state to ensure the results are the same each time the code is run)
```

### 2.3.2. Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
# Initializing the Regression Model (we start with empty data)
lr = LinearRegression()  
```

```python
# Data training
lr.fit(X_train, y_train)   # Variables with 70% for training
```

```python
# Data testing
prediction_lr = lr.predict(X_test)   # predict (predicts Values_ha (y = output) having the rest of variables as characteristics of the Values_ha (x = input)
```

```python
# Results of the prediction (predict variable y)
prediction_lr
```

## 2.4.  Performance Metric

To know the accuracy of the model

```python
# Measures the distance between y(actual) and y(predicted)
# Measures the average distance from the real points with the predictive points
# Compare the predicted with the actual value to know if the model is correct

from sklearn.metrics import r2_score
r2_lr = r2_score(y_test, prediction_lr)  
r2_lr
```

There are other metrics for regression

Note: These metrics depend on the scale of the data!

- MSE (Mean Squared Error): average of the squared difference between the real value and the predicted one. (penalty for error)
- MAE (Mean Absolute Error): average of the absolute difference (module) between the real value and the predicted one.

# 5. Performance Techniques

- Optimize the model and obtain the better results
- I'll choose the Decision Tree and Random Forest models.

## Optimization Tools

### A) Cross Validation 

- Cross Validation is a widely used technique to evaluate the performance of models and subsequent verification of trends in machine learning. It consists of partitioning the data into subsets (parts), where one subset is used for training and another subset is used for testing and evaluating the performance of the model
- The partitioned data is the training data (70% of the original data)
- K-fold: the training set is randomly divided into K subsets of approximately the same size, then K iterations are performed, where K-1 subsets are for training and 1 for validation (number of partitions - subsets) (k subsets , k tests)
- Performance calculated by k-fold cross validation is the average of all values calculated in each iteration

### B) Randomized Search

- "Defines a search space as a bounded domain of hyperparameter values and random sample points in that domain" (searches for the best hyperparameters of the model and chooses possible combinations at random)

### C) Grid Search

- Make all possible combinations between hyperparameters
- It takes a longer computational time to obtain the results

## 5.1. Optimize Decision Tree

```python
# Division of the data into 5 pieces
from sklearn.model_selection import KFold
```

```python
# Define the number of subsets K
cv = KFold(n_splits=5, shuffle=True)    # shuffle (choose random numbers)
```

### 5.1.1. Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
# The average of the 5 executions is printed to know the final precision (accuracy)
accuracy_dtr_cross_val_r2 = cross_val_score(estimator=dtr, X=X_train, y=y_train, scoring='r2', cv=cv)    # 5 (r2) will be obtained since we have 5 partitions and the (r2) will be the average of all
print('Average Accuracy of r2: {}'.format(round(accuracy_dtr_cross_val_r2.mean(),3)))     # r2 = Determination Coefficient
```

### 5.1.2. Randomized Search 

```python
# Selection of Hyperparameters
# Make all possible combinations randomly
param_random_dt = {
'criterion': ['squared_error', 'absolute_error'],  # Divide the nodes trying to reduce the error using these criteria
'max_depth': [6, 8, 10, 12, 14, 16],               # Tree depth, number of stories
'min_samples_leaf': [6, 8, 10],                    # Number of sheets, final result, price of the property
'min_samples_split': [8, 10, 12, 14, 16],          # Decision node, decision making quantity
'max_features': ['sqrt', 'log2', 'auto'],}         # Find node split precision
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Inizializing
random_cv_dt = RandomizedSearchCV(dtr, param_random_dt, n_iter=5, cv=cv, scoring='r2', n_jobs=-1)   # n_jobs=-1 (low value so that the machine does not collapse)
```

```python
randomcv_dtr = random_cv_dt.fit(X_train, y_train)    # It takes time to process this code as it processes the possible combinations
```

```python
pd.DataFrame(randomcv_dtr.cv_results_)
```

```python
# The best combination
randomcv_dtr.best_params_
```

```python
# The best score
randomcv_dtr.best_score_
```

```python
# Create a bar chart to display the predictions

years = predictions_df['Year'].tolist()
predicted_values = predictions_df['Predicted_Value_ha'].tolist()        # Create a line plot to display the predictions

plt.figure(figsize=(12, 6))
plt.plot(years, predicted_values, marker='o', linestyle='-')
plt.title('Afforestation Predicted by Year (2023-2030)', fontsize=20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Predicted_Value_ha', fontsize=15)
plt.xticks(rotation=0)
plt.show()
```
