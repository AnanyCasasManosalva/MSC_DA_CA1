## 1.3. Debugging the data

#### I'll delete the rows that contain 'Total Afforestation' since I want to analyse the plantations carried out for each specie and forest owner instead of analysing them in their entirety, this is to avoid generating redundancy or noise

```python
filtro = (afforestation_df['Species'] != 'Total Afforestation') & (afforestation_df['Forest Owner'] != 'Total Afforestation')    # Filter the DataFrame to remove rows where the columns 'Species' or 'Forest Owner' contain 'Total Afforestation'
afforestation_df = afforestation_df[filtro]   # Update the DataFrame with the requested conditions 
afforestation_df                              # Display the DataFrame
```

#### I'll remove columns 'Statistic Label' and 'UNIT' due to they both have the same value: 'Afforestation Area' and 'Hectares' respectively

```python
afforestation_df.drop(['Statistic Label', 'UNIT'], axis=1, inplace=True)   # Drop (method to delete rows or columns in a DataFrame), axis=1 (remove columns instead of rows), inplace=True (modify the DataFrame in place)
```

#### We'll rename the columns 'VALUE' and 'Forest Owner'

```python
column_dictionary = {'VALUE':'Value_ha', 'Forest Owner':'Forest_Owner'}         # Creation of a dictionary
afforestation_df.rename(columns=column_dictionary, inplace=True)                # Rename (method to rebrand columns named 'VALUE' to 'Value_ha' and 'Forest Owner' to 'Forest_Owner'), inplace=True (replace the new changes in the DataFrame)
afforestation_df                                                                # Display the DataFrame to see the changes made
```

## Binomial Distribution

#### This distribution models success or failure events and is used for discrete data.
#### If I apply Binomial Distribution my probability of success would be 'planting trees' and failure 'not planting trees', in a specific area measured in hectares, which would be an unrealistic case, but let's see what the distribution looks like.

```python
# Binomial Distribution

from scipy.stats import binom
import matplotlib.pyplot as plt

n_ensayos = len(afforestation_df)  # Número total de ensayos (filas en tus datos)
prob_exito = 0.5  # Probabilidad de éxito (puedes ajustar esto según tus datos)
binomial_dist = binom(n_ensayos, prob_exito) # Generar la distribución binomial
plt.hist(afforestation_df['Value_ha'], bins=20, density=True, alpha=0.5, color='blue', label='Real data') # Histograma de datos reales

x = range(int(afforestation_df['Value_ha'].min()), int(afforestation_df['Value_ha'].max()) + 1)  # Función de masa de probabilidad para la distribución binomial
plt.plot(x, binomial_dist.pmf(x), 'r-', label='Binomial PMF')

plt.title('Fit: Real Data vs. Binomial Distribution')
plt.xlabel('Value_ha')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

#### It can be observed that the Real Distribution differs from the Binomial Distribution, so it is not convenient to use it.

## Poisson Distribution

#### This distribution is used to model events that occurred in a time interval. 
#### In the case of my data, I could model the tree planting rate per year.

```python
# Poisson Distribution

from scipy.stats import poisson
import matplotlib.pyplot as plt

tasa = afforestation_df['Value_ha'].mean()        # Tasa de ocurrencia de eventos (ajusta según tus datos)
poisson_dist = poisson(tasa)                      # Generar la distribución de Poisson
plt.hist(afforestation_df['Value_ha'], bins=20, density=True, alpha=0.5, color='blue', label='Real data')  # Histograma de datos reales

x = range(int(afforestation_df['Value_ha'].min()), int(afforestation_df['Value_ha'].max()) + 1) # Función de masa de probabilidad para la distribución de Poisson
plt.plot(x, poisson_dist.pmf(x), 'r-', label='Poisson PMF')

plt.title('Fit: Real Data vs. Poisson Distribution')
plt.xlabel('Value_ha')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

#### It can be seen that both distributions are similar and follow the same trend, although the Poisson Distribution presents larger peaks than the Real Distribution.
#### Later we will make use of the Normal Distribution to decide which of them will be suitable for generating prediction models

### Transformation of Variables

```python
# FunctionTransformer transforms real values to transformed values using logarithmic function
# To return to the initial values, we transform the logarithmic function into an exponential function (inversely proportional)

from sklearn.preprocessing import FunctionTransformer
transformer = FunctionTransformer(np.log1p, validate=True)  # Np.log1p (creates a transformer object that applies the 'log1p' function from the NumPy library, 'validate=True' (argument ensures that the transformation is validated)
transformed_values = transformer.transform(afforestation_df.select_dtypes(exclude=['object']))  # Transform (method to transform numeric columns, excluding object data types, into logarithmic function)
```

```python
columns_transformed_values = afforestation_df.select_dtypes(exclude=['object']).columns      # Select_dtypes(exclude=['object']).columns (method to select the columns from the DataFrame that don't have an object data type)
transformed_df = pd.concat([afforestation_df.select_dtypes(include=['object']), pd.DataFrame(transformed_values, columns=columns_transformed_values)], axis=1) # Concat (method to join two DataFrames along the columns or rows (in this case columns). This line combines the categorical columns with the transformed numerical columns) 
transformed_df.head()           # Head (method to show the first 5 rows of the DataFrame)
```

```python
# After the transformation

sns.heatmap(transformed_df.corr(), annot = True, fmt = '.2f', linewidths = 2)    # Use the Seaborn library to generate Heatmaps in colored shapes, which shows the correlation between the numeric variables of the DataFrame, annot=True (show the values within each cell of the Heatmap cell), fmt='.2f' (display the values in cells with two decimal places), linewidth=2 (set the linewidth of edges to 2)
ax.set_title('Correlation between Variables - Transformation Log', fontsize=20)  # Set the graph title as "Correlation between Variables - Transformation Log" with a font size of 20
plt.show()                                                                       # Display the Heat Map 
```

# 2.Linear Regression

"Regression analysis refers to the study of the dependence of a variable, the dependent variable, in relation to one or more variables, the explanatory variables, with the objective of estimating and/or predicting the mean value (of the population) of the first in terms of the known or fixed values (in repeated sampling) of the latter.”

- Simple regression: the dependent variable (y) is related to a single explanatory variable (X).
- Multiple regression: The dependent variable (y) is related to more than a single explanatory variable (Xi).

## 2.1. Adjustment and prediction

### 2.1.1. Data division

```python
# Identify the variables that will be dependent (y) and independent (X)

# y: variable that I'm going to predict (Value_ha)
# X: variables as characteristics of the Value_ha

# Independent Variable
X = afforestation_dummies_df.drop(['Value_ha'], axis=1) # Variable that has the characteristics of the variable ‘y’

# Dependent Variable
y = afforestation_dummies_df['Value_ha']                # Variable that I'm going to predict
```

```python
y      # Display the dependent variable
```

```python
from sklearn.model_selection import train_test_split    # Import the train_test_split class from the model.selection module of the scikit-learn (sklearn) library
```

```python
# Splitting the data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)     # train_test_split (function used to split a dataset into two or more subsets,test_size=0.3 (use 30% of the data for testing and 70% for training), random_state=42 (set a specific value for random_state to ensure the results are the same each time the code is run)
```

### 2.1.2. Linear Regression Model

```python
from sklearn.linear_model import LinearRegression   # Import the LinearRegression class from the linear_model module of the scikit-learn (sklearn) library
```

```python
lr = LinearRegression()  # Initializing the Linear Regression Model (we start with empty data) 
```

```python
# Data training
lr.fit(X_train, y_train)   # X_train, y_train will be the variables that represent 70% of the data to be evaluated for the training of the Linear Regression Model 
```

```python
# Data testing
prediction_lr = lr.predict(X_test)   # lr.predict (predicts values 'y' (output) having another variable, called 'X' (input) as characteristics of 'y')
```

```python
# Results of the prediction (predict variable y)
prediction_lr   # Display the results of the prediction (variable ‘y’) using a trained Linear Regression Model (lr) on a set of input features (X)
```

```python
# Inverse transformation (convert a logarithmic number to an exponential)
np.expm1(4.42089844e+00)
```

## 2.2.  Performance Metric

To know the accuracy of the model

```python
# Measure the distance between y(actual) and y(predicted)
# Measure the average distance from the real points with the predictive points
# Compare the predicted with the actual value to know if the model is correct

from sklearn.metrics import r2_score        # imports the R-squared scoring function from the sklearn.metrics module
r2_lr = r2_score(y_test, prediction_lr)     # The r2_score function compares how well the predictions (prediction_lr) match the actual values (y_test)
r2_lr                                       # Display the Determination Coeficient (r2)
```

We're going to use r2 score as metrics but there are other metrics for regression

Note: These metrics depend on the scale of the data!

- MSE (Mean Squared Error): average of the squared difference between the real value and the predicted one. (penalty for error)
- MAE (Mean Absolute Error): average of the absolute difference (module) between the real value and the predicted one.

# 5. Performance Techniques

- Optimize the model and obtain the better results
- I'll choose the Decision Tree and Random Forest models.

## Optimization Tools

### A) Cross Validation 

- Cross Validation is a widely used technique to evaluate the performance of models and subsequent verification of trends in machine learning. It consists of partitioning the data into subsets (parts), where one subset is used for training and another subset is used for testing and evaluating the performance of the model
- The partitioned data is the training data (70% of the original data)
- K-fold: the training set is randomly divided into K subsets of approximately the same size, then K iterations are performed, where K-1 subsets are for training and 1 for validation (number of partitions - subsets) (k subsets , k tests)
- Performance calculated by k-fold cross validation is the average of all values calculated in each iteration

### B) Randomized Search

- "Defines a search space as a bounded domain of hyperparameter values and random sample points in that domain" (searches for the best hyperparameters of the model and chooses possible combinations at random)

### C) Grid Search

- Make all possible combinations between hyperparameters
- It takes a longer computational time to obtain the results

## 5.1. Optimize Decision Tree

```python
# Division of the data into 5 pieces
from sklearn.model_selection import KFold  # Import KFold class from the scikit-learn library
```

```python
# Define the number of subsets K
cv = KFold(n_splits=5, shuffle=True)  # n_splits=5 (split the data into 5 pieces), shuffle=True (choose random numbers)
```

### 5.1.1. Cross Validation

```python
from sklearn.model_selection import cross_val_score   # Import cross_val_score class from the scikit-learn library
```

```python
# The average of the 5 executions is printed to know the final precision (accuracy)
# This function performs a cross-validation scoring of the estimator (dtr) on the input data (X and y training)
accuracy_dtr_cross_val_r2 = cross_val_score(estimator=dtr, X=X_train, y=y_train, scoring='r2', cv=cv)    # estimator=dtr (specifies the Decision Tree Regressor model to use), scoring='r2' (5 (r2) will be obtained since we have 5 partitions and the (r2) will be the average of all)   
print('Average Accuracy of r2: {}'.format(round(accuracy_dtr_cross_val_r2.mean(),3)))     # r2 = Determination Coefficient
```

### 5.1.2. Randomized Search 

```python
# Selection of Hyperparameters
# Make all possible combinations randomly
param_random_dt = {
'criterion': ['squared_error', 'absolute_error'],  # Divide the nodes trying to reduce the error using these criteria
'max_depth': [6, 8, 10, 12, 14, 16],               # Tree depth, number of stories
'min_samples_leaf': [6, 8, 10],                    # Number of sheets, final result, afforestation in hectares
'min_samples_split': [8, 10, 12, 14, 16],          # Decision node, decision making quantity
'max_features': ['sqrt', 'log2', 'auto'],}         # Find node split precision
```

```python
from sklearn.model_selection import RandomizedSearchCV  # Import RandomizedSearchCV class from the scikit-learn library
```

```python
# Inizializing
random_cv_dt = RandomizedSearchCV(dtr, param_random_dt, n_iter=5, cv=cv, scoring='r2', n_jobs=-1)   # n_iter=5 (number of random combinations of hyperparameters to try), cv=cv (cv parameter is a cross-validation splitting), scoring='r2 (use the R-squared score as the evaluation metric), n_jobs=-1 (low value so that the machine does not collapse)
```

```python
# X_train, y_train will be the variables that represent 70% of the data to be evaluated for the training of the Random Forest Regressor Model
randomcv_dtr = random_cv_dt.fit(X_train, y_train)    # It takes time to process this code as it processes the possible combinations
```

```python
pd.DataFrame(randomcv_dtr.cv_results_)     # Display the values obtained from the possible combinations made
```

```python
# The best combination
randomcv_dtr.best_params_     # Display the best hyperparameters found during the randomized search
```

```python
# The best score
randomcv_dtr.best_score_    # Display the best estimator of the mean cross-validated score
```
